```python
from pathlib import Path
import nibabel as nb 
from pycimg import CImg
import matplotlib.pyplot as plt

data_dir= Path('data')


def get_CT_scans():
    idxs = [32,36,39]
    filenames = [data_dir/f'MASK_00{idx}.nii.gz' for idx in idxs]
    return [nb.load(file).get_fdata() for file in filenames]
```

```python
scans = get_CT_scans()
```

```python
CImg(scans[0]).display()
```

```python

x,y,z = scans[0].shape

fig, subplots = plt.subplots(nrows=13, ncols=10, figsize=(20,20)) 
for slice_idx, axs in enumerate([plt for row in subplots for plt in row]):
    axs.axis('off')
    if slice_idx < z:
        axs.imshow(scans[0][:,:,slice_idx], cmap='gray')
```
